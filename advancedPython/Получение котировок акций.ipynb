### Задание: получение котировок акций
Получите данные по котировкам акций со страницы:\
https://mfd.ru/marketdata/?id=5&group=16&mode=3&sortHeader=name&sortOrder=1&selectedDate=01.11.2019 \
и найдите, по какому тикеру был максимальный рост числа сделок (в процентах) за 1 ноября 2019 года.

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

r = requests.get("https://mfd.ru/marketdata/?id=5&group=16&mode=3&sortHeader=name&sortOrder=1&selectedDate=01.11.2019")
html = BeautifulSoup(r.content)
table = html.find('table', {'id':'marketDataList'})
rows = []
trs = table.find_all('tr')
for tr in trs:
    tr = [td.get_text(strip=True) for td in tr.find_all('td')]
    if len(tr) > 0:
        rows.append(tr)
        
# Собрали данные с web-страницы
data = pd.DataFrame(rows )
# data.head()
# Переименовываем колонки
columns=["Тикер", "Время", "Сделки", "C1", "С%", "Закрытие", "Открытие", "min", "max", "avg", "шт", "руб", "Колво_сделок"]
data.columns=columns
```

```python
data = data[data["Сделки"] != "N/A"]   # подсказано из решения
data['С%'] = data['С%'].str.replace("%","").str.replace("−","-").astype(float)  # иначе нет конверсии отр.чисел
data.head()  #.info()
```

```python
# print(data.sort_values('С%',ascending=False).head(1))

print("максимальный рост числа сделок (в процентах) за 1 ноября 2019 года был у")
print(data.sort_values('С%',ascending=False).head(1).Тикер)
```
